## <font color="green">4) Sets</font>

Sets automatically remove duplicates and are optimized for **fast membership testing** (`in`).

Sets are commonly used for:

```python
# Creating a set (duplicates are removed automatically)
nums = {1, 2, 2, 3}
print(nums)          # {1, 2, 3}

# Adding and removing items
nums.add(4)
nums.remove(2)       # raises KeyError if 2 is not present
nums.discard(99)     # does nothing if 99 is not present

# Membership testing (fast)
print(3 in nums)

# Set operations
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)  # union -> {1, 2, 3, 4, 5}
print(a & b)  # intersection -> {3}
print(a - b)  # difference -> {1, 2}
```

```python
# Creating a DataFrame
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "Dina", "Evan"],
    "Age": [25, 30, 35, 28, 32],
    "Score": [88, 92, 95, 85, 90]
})

df
```

### DataFrame Operations

We can filter, sort, and summarize data efficiently.


```python
# Generating summary statistics for the numeric columns in the DataFrame df
print(df.describe())

# Filtering rows based on a condition and Sorting the filtered results
top = df[df["Score"] >= 90].sort_values("Score", ascending=False)

# Displaying the result
top
```

### Visualization: Pandas DataFrame

Bar and scatter plots help reveal patterns in the data.


```python
# Bar Chart (Scores by Student)
plt.figure()
plt.bar(df["Name"], df["Score"])
plt.title("Scores by Student")
plt.ylabel("Score")
plt.ylim(0, 100)
plt.grid(axis="y")
plt.show()

# Scatter Plot (Age vs Score)
plt.figure()
plt.scatter(df["Age"], df["Score"])
plt.title("Age vs Score")
plt.xlabel("Age")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```
